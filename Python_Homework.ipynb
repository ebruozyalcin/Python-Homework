```python
print("__________Student Management System__________")

print("\n__Registering students to the system__\n")

studentlist = []

for i in range(3):

    student_n1 = input("Please enter your name: ")
    student_s1 = input("Please enter your surname: ") 

    studentlist.insert(i + 1, student_n1 + " " + student_s1)
    
print("\nList of registered students : {}\n" .format(studentlist))

print("\n__Login Screen__\n")

student_n = "Ebru"
student_s = "Özyalçın"

check = False

for x in range(3):
    student_n1 = input("Please enter your name: ")
    student_s1 = input("Please enter your surname: ") 
    if (student_n == student_n1 and student_s == student_s1):
        print("Welcome")
        check = True
        break
    elif (student_n != student_n1 and student_s == student_s1):
        print("You entered your name wrong...")
    elif (student_n == student_n1 and student_s != student_s1):
        print("You entered your surname wrong...")
    elif (student_n != student_n1 and student_s != student_s1):
        print("You entered your name and surname wrong...")
    
if not check:
    print("Please try again later..")

if (check == True):
    print("\n__Taken courses__\n")
    
    courselist = []
    
    course_numbers = int(input("Enter the number of courses you want to take: "))
    
    def message():
        return "You failed in class!!"


    i = 0

    if 3 <= course_numbers <= 5:
        while(i < course_numbers):
            Coursename = input('Enter the course name: ')
            courselist.append(Coursename)
            i += 1
            
    elif course_numbers > 5:
        print("You cannot take more than 5 courses!")
    else:
        print(message())
        
    
    print("Taken Courses: {}".format(courselist)) 


    if len(courselist) > 2:
        print("\n__Exam Notes for {} Course__\n".format(courselist[0]))
        midterm = int(input("Please enter the midterm grade: ")) 
        final = int(input("Please enter the final grade: "))
        project = int(input("Please enter the project grade: "))
    
        grades = {"Midterm": midterm, "Final": final, "Project": project}
        
        if (0 <= midterm <= 100 and 0 <= final <= 100 and 0 <= project <=100):
            print("\nGrades: {}\n".format(grades))
            
            midterm_grade = midterm * 0.30
            final_grade = final * 0.5
            project_grade = project * 0.2

            Grade = midterm_grade + final_grade + project_grade

            print("The student's grade is {}\n".format(Grade, ".2f"))

            if 90 < Grade <= 100:
                print("The student's note is: AA")
            elif 70 < Grade <= 90:
                print("The student's note is: BB")
            elif 50 < Grade <= 70:
                print("The student's note is: CC")
            elif 30 < Grade <= 50:
                print("The student's note is: DD")
            elif Grade < 30:
                print("The student's note is: FF\n\nSince the student takes an FF note, failed this course!!")
        else:
            print("You entered invalid grade!")
        
          
    
```
